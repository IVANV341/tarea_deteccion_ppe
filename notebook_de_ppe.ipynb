```python
import os

dataset_path = "/content/drive/MyDrive/ppe/ppe-factory-1"

for root, dirs, files in os.walk(dataset_path):
    level = root.replace(dataset_path, "").count(os.sep)
    indent = "  " * level
    print(f"{indent}📁 {os.path.basename(root)}/")
    if level < 2:
        subindent = "  " * (level + 1)
        for file in files[:5]:
            print(f"{subindent}📄 {file}")
        if len(files) > 5:
            print(f"{subindent}... y {len(files) - 5} archivos más")
```

```python
import os

train_images = "/content/drive/MyDrive/ppe/ppe-factory-1/train/images"
train_labels = "/content/drive/MyDrive/ppe/ppe-factory-1/train/labels"

imagenes = set(os.path.splitext(f)[0] for f in os.listdir(train_images))
etiquetas = set(os.path.splitext(f)[0] for f in os.listdir(train_labels))

print(f"📷 Imágenes:  {len(imagenes)}")
print(f"🏷️  Etiquetas: {len(etiquetas)}")
print(f"{'✅ Todas las imágenes tienen su etiqueta' if imagenes == etiquetas else '⚠️ Hay inconsistencias'}")

sin_etiqueta = imagenes - etiquetas
sin_imagen   = etiquetas - imagenes

if sin_etiqueta:
    print(f"\n❌ Imágenes SIN etiqueta ({len(sin_etiqueta)}):")
    for f in list(sin_etiqueta)[:10]:
        print(f"  - {f}")

if sin_imagen:
    print(f"\n❌ Etiquetas SIN imagen ({len(sin_imagen)}):")
    for f in list(sin_imagen)[:10]:
        print(f"  - {f}")
```

```python
from ultralytics import YOLO

model = YOLO("yolo11n.pt")

results = model.train(
    data="C:/ppe_dataset/data.yaml",
    epochs=50,
    imgsz=640,
    batch=16,
    project="C:/ppe_dataset/runs",
    name="entrenamiento_ppe",
    device=0
)

print(f"✅ Entrenamiento completado")
print(f"📁 Resultados en: C:/ppe_dataset/runs/entrenamiento_ppe")
```

tiempo de entrenamiento: aprox 50 minutos


```python
import os
from ultralytics import YOLO

model = YOLO("C:/ppe_dataset/runs/entrenamiento_ppe/weights/best.pt")

metrics = model.val(data="C:/ppe_dataset/data.yaml")

print(f"\n📊 Resultados finales:")
print(f"mAP50:    {metrics.box.map50:.3f}")
print(f"mAP50-95: {metrics.box.map:.3f}")
print(f"Precision:{metrics.box.mp:.3f}")
print(f"Recall:   {metrics.box.mr:.3f}")
```

```python
from ultralytics import YOLO

model = YOLO("C:/ppe_dataset/runs/entrenamiento_ppe/weights/best.pt")

results = model.predict(
    source="C:/ppe_dataset/valid/images",
    save=True,
    conf=0.5,
    project="C:/ppe_dataset/runs",
    name="inferencia_ppe"
)

print(f"✅ Inferencia completada")
print(f"📁 Imágenes guardadas en: C:/ppe_dataset/runs/inferencia_ppe")
```
